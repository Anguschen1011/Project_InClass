```python
data = [
    ['A','B','C','D'],
    ['A','B','C','E'],
    ['B','C'],
    ['B','E']
]
print(type(data))
print(data)
```

```python
from mlxtend.preprocessing import TransactionEncoder

TE = TransactionEncoder()
X = TE.fit(data)
print(X)

TDB = TE.fit(data).transform(data)

print(type(TDB))
print(TDB)
```

```python
Item_list = TE.columns_
print(type(Item_list))
print(Item_list)
```

```python
DB = TDB.astype("int")
print(type(DB))
print(DB)
```

```python
import pandas as pd

df = pd.DataFrame(DB, columns = Item_list)

print(df)
```

```python
from mlxtend.frequent_patterns import apriori

frequent_item = apriori(df, min_support=0.5, use_colnames=True)

print(frequent_item)
```

```python
from mlxtend.frequent_patterns import association_rules

rule = association_rules(frequent_item, metric="confidence", min_threshold=0.6)

print(rule)
rule.to_csv("Rule.csv")
```
