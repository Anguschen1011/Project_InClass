```python
from sklearn.datasets import make_blobs

X,_ = make_blobs(n_samples = 25, n_features = 2, centers = 3, cluster_std = 1)

print(X)
print(type(X))
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0], X[:,1], color = "blue" )

#plt.scatter([1,2,3],[4,5,6],color = "red")
```

```python
from sklearn.cluster import AgglomerativeClustering as ac

model = ac(n_clusters = 3, linkage = "single")

model.fit(X)

model_pred = model.fit_predict(X)

print(type(model_pred))
print(model_pred.shape)
print(model_pred)
```

```python
from scipy.spatial.distance import pdist

xdis = pdist(X, metric = "cityblock")

print(xdis)
```

```python
from scipy.cluster.hierarchy import linkage

lr = linkage(xdis, method = "single")

print(lr)
```

```python
from scipy.cluster.hierarchy import dendrogram

R = dendrogram(lr)

print(R)
```

```python
from sklearn import metrics

avg_score = metrics.silhouette_score(X, model_pred)

print(avg_score)
```

```python
each_score=  metrics.silhouette_samples(X, model_pred)

print(each_score)
```

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# 輸入距離矩陣
x = np.array([
    [0, 7, 9, 3, 8],
    [7, 0, 2, 6, 4],
    [9, 2, 0, 11, 10],
    [3, 6, 11, 0, 5],
    [8, 4, 10, 5, 0]
])

# 使用 Single Linkage 方法進行層次聚類
single_linkage = linkage(x, method='single', optimal_ordering=True)

# 使用 Complete Linkage 方法進行層次聚類
complete_linkage = linkage(x, method='complete', optimal_ordering=True)

# 設置子圖並繪製階層圖
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# 繪製 Single Linkage 階層圖
dendrogram(single_linkage, ax=axes[0])
axes[0].set_title('Single Linkage Dendrogram')
axes[0].set_xlabel('Data Points')
axes[0].set_ylabel('Distance')

# 繪製 Complete Linkage 階層圖
dendrogram(complete_linkage, ax=axes[1])
axes[1].set_title('Complete Linkage Dendrogram')
axes[1].set_xlabel('Data Points')
axes[1].set_ylabel('Distance')

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

# 假设您的数据
data = np.array([[2, 2], [8, 9], [1, 3], [9, 10], [3, 4], [10, 11]])

# 初始化群C1和C2的中心点
initial_centers = np.array([[2, 2], [10, 11]])

# 使用 KMeans 进行分群
kmeans = KMeans(n_clusters=2, init=initial_centers, n_init=1)
kmeans.fit(data)

# 输出分群结果
print("Cluster centers:\n", kmeans.cluster_centers_)
print("Cluster labels:", kmeans.labels_)
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

# 假设您的距离矩阵
distance_matrix = np.array([[0, 2.5, 5, 7.8, 9.4],
                            [2.5, 0, 3, 5.8, 8.4],
                            [5, 3, 0, 3.3, 4.9],
                            [7.8, 5.8, 3.3, 0, 2.1],
                            [9.4, 8.4, 4.9, 2.1, 0]])


# 定义 Eps 和 MinPts
eps = 3.5
min_samples = 3

# 创建 DBSCAN 模型并进行拟合
dbscan = DBSCAN(eps=eps, min_samples=min_samples, metric='precomputed')
labels = dbscan.fit_predict(distance_matrix)

# 输出分群结果
print("Cluster labels:", labels)
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

# 假设您的距离矩阵
distance_matrix = np.array([[0, 1, 6, 4, 4.4],
                            [1, 0, 5.2, 3.2, 3],
                            [6, 5.2, 0, 2.6, 1.7],
                            [4, 3.2, 2.6, 0, 1.4],
                            [4.4, 3, 1.7, 1.4, 0]])

# 定义 Eps 和 MinPts
eps = 1.7
min_samples = 3

# 创建 DBSCAN 模型并进行拟合
dbscan = DBSCAN(eps=eps, min_samples=min_samples, metric='precomputed')
labels = dbscan.fit_predict(distance_matrix)

# 获取核心点、边界点和噪声点的索引
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True

# 输出分群结果和点的类型
for i, label in enumerate(labels):
    if label == -1:
        print(f"Point {i} is a Noise Point")
    elif core_samples_mask[i]:
        print(f"Point {i} is a Core Point in Cluster {label}")
    else:
        print(f"Point {i} is a Border Point in Cluster {label}")
```
